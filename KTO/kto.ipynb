## 데이터 전처리 과정 함수 정의
(위에서 한 것 함수로 정의)
- excel 불러오기
    - 년과 월 입력받아서 (1 넣어도 01월 나오도록)
- 년월 컬럼 추가
- 대륙 제거하고 국가만 남기기
- 대륙 컬럼 추가
- 관광객 비율
- 전체 비율

```python
def create_kto_data(yy,mm):
    
    #엑셀 파일 불러오기
    mm=str(mm).zfill(2)
    path=f"files/kto_{yy}{mm}.xlsx"
    df=pd.read_excel(path, header=1,
                 skipfooter=4, usecols='A:G')
    
    #년월 열 추가
    df['기준년월']=f"{yy}-{mm}"
    
    #대륙 제거(국가만 남기기) 후 인덱스 재설정
    ignore_list=['아시아주','미주','구주','대양주','아프리카주','기타대륙','교포소계']
    cond=df['국적'].isin(ignore_list)==False
    df_country=df[cond].reset_index(drop=True)
    
    #대륙 열 추가
    #대륙 컬럼 생성을 위한 목록 만들어 놓기 
    continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 + ['대양주']*3 + ['아프리카']*2 + ['기타대륙'] + ['교포']    
    df_country['대륙'] = continents
    
    #관광객 비율
    df_country['관광객비율(%)'] = round(df_country.관광 / df_country.계 * 100, 1)

    #전체 비율
    tourist_sum = sum(df['관광'])
    df_country['전체비율(%)'] = round(df_country['관광'] / tourist_sum * 100, 1)
    df_country.sort_values('전체비율(%)', ascending=False)
    
    return(df_country)
```

```python
create_kto_data(2015,8)
```

```python
df=pd.DataFrame()
```

```python
for yy in range(2010, 2021):
    for mm in range(1, 13):
        temp = create_kto_data(yy,mm)
        df=df.append(temp, ignore_index=True)
```
