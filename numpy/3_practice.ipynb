# concat 메서드 사용하기

```python
import pandas as pd

df1 = pd.read_csv('../../Downloads/pandas_data/concat_1.csv') 
df2 = pd.read_csv('../../Downloads/pandas_data/concat_2.csv') 
df3 = pd.read_csv('../../Downloads/pandas_data/concat_3.csv')
```

```python
print(df1)
print(df2)
print(df3)
```

```python
pd.concat([df1,df2,df3])
```

행인덱스 번호 이어지게 하려면 원래 데이터의 인덱스를 무시해야함

```python
pd.concat([df1,df2,df3], ignore_index=True)
```

```python
pd.concat([df1,df2,df3],axis=1)
```

데이터프레임에 딕셔너리 붙이기

# merge 메서드 사용하기
