# concat 메서드 사용하기

```python
import pandas as pd

df1 = pd.read_csv('../../Downloads/pandas_data/concat_1.csv') 
df2 = pd.read_csv('../../Downloads/pandas_data/concat_2.csv') 
df3 = pd.read_csv('../../Downloads/pandas_data/concat_3.csv')
```

```python
print(df1)
print(df2)
print(df3)
```

```python
pd.concat([df1,df2,df3])
```

```python
pd.concat([df1,df2,df3],axis=1)
```

# merge 메서드 사용하기

```python
person = pd.read_csv('../../Downloads/pandas_data/survey_person.csv') 
site = pd.read_csv('../../Downloads/pandas_data/survey_site.csv') 
survey = pd.read_csv('../../Downloads/pandas_data/survey_survey.csv') 
visited = pd.read_csv('../../Downloads/pandas_data/survey_visited.csv')
```

```python
person
```

```python
survey
```

```python
pd.merge(survey,person,left_on='person',right_on='ident',how='right')
```

```python
visited
```

```python
site
```

```python
pd.merge(visited,site,left_on='site',right_on='name')
```
